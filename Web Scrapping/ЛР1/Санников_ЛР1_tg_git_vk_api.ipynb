```python
from telethon import TelegramClient, events, sync
from telethon.tl.functions.channels import GetFullChannelRequest
```

```python
async def parseChatMessages(session_name, api_id, api_hash, phone, chat_link, limit = 100):
        client = TelegramClient(session_name, api_id, api_hash)

        # Get the chat entity
        await client.start(phone=phone)
        try:
          me = print(await client.get_me())
        except:
          print("not me")
          await client.disconnect()
        try:
          chat_entity = await client.get_entity(chat_link)
          full_ch = await client(GetFullChannelRequest(channel=chat_entity))
          messages = await client.get_messages(chat_entity, limit=limit)
          print(chat_entity.title, "(",full_ch.full_chat.participants_count,")")

          parsed_messages = []
          for i in messages:
            parsed_messages.append({
                'id':i.id,
                'date': i.date,
                'sender_id': i.sender_id,
                'text': i.text,
                'media': i.media
            })
          return parsed_messages
        finally:
          await client.disconnect()
```

```python
async def parseChanelMessages(session_name, api_id, api_hash, phone, chat_link, limit = 100):
        client = TelegramClient(session_name, api_id, api_hash)

        # Get the chat entity
        await client.start(phone=phone)
        try:
          chat_entity = await client.get_entity(chat_link)
          full_ch = await client(GetFullChannelRequest(channel=chat_entity))
          messages = await client.get_messages(chat_entity, limit=limit)
          print(chat_entity.title, "(",full_ch.full_chat.participants_count,")")

          parsed_messages = []
          for i in messages:
            # if not i.message in messages:
            #   continue
            message_data = ({
                'id':i.id,
                'date': i.date,
                'sender_id': i.sender_id,
                'text': i.text,
                'media': i.media,
                'replies': i.replies,
                'is_post': getattr(i, 'post', False),
                'views': i.views,
                'reactions': getattr(i, 'reactions', None),
            })
            parsed_messages.append(message_data)
          return parsed_messages
        except Exception as e:
          print("error", e)
          return[]
        finally:
          await client.disconnect()
```

```python

# Replace with your GitHub username and Personal Access Token
GITHUB_USERNAME = "Uranus28"
GITHUB_PAT = first_line

# Endpoint to list repositories for the authenticated user
url = f"https://api.github.com/users/{GITHUB_USERNAME}/repos"

headers = {
    "Authorization": f"token {GITHUB_PAT}",
    "Accept": "application/vnd.github.v3+json"
}
```

```python

try:
    response = requests.get(url, headers=headers)
    response.raise_for_status()  # Raise an exception for HTTP errors

    repos_data = response.json()

    for repo in repos_data:
        print(repo)
        print(f"Repo Name: {repo['name']}")
        print(f"Description: {repo['description']}")
        print(f"HTML URL: {repo['html_url']}")
        print("-------------------------")

except requests.exceptions.RequestException as e:
    print(f"Error :((( {e}")
```
